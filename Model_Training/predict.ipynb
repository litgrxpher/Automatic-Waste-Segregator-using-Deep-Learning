```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image
import numpy as np
# Load the pre-trained model
model_file = "AWS_CNN.h5"
model = load_model(model_file)

# Example: Load and preprocess an image
image_path = "metal.png"  # Replace with the path to your image file
image = Image.open(image_path)
image = image.resize((300, 300))  # Adjust the size according to your model's input size
image_array = np.array(image) / 255.0  # Normalize pixel values to the range [0, 1]
input_data = np.expand_dims(image_array, axis=0)  # Add batch dimension

# Make predictions
predictions = model.predict(input_data)

# Get class labels
class_labels = ["metal", "organic", "paper", "plastic"]

# Decode predictions
predicted_class_index = np.argmax(predictions[0])
predicted_class = class_labels[predicted_class_index]

# Display the image
plt.imshow(image)
plt.axis('off')
plt.title(f'Predicted class: {predicted_class}')
plt.show()
```
